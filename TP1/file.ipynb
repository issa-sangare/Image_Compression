```python
from IPython.display import display
from PIL import Image
import numpy as np

image_file = Image.open("./img/eren_rgb.png") # open colour image
image_file = image_file.convert('1') # convert image to black and white
image_file.save('./img/eren_rgb_bin.png')
display(image_file)
print(np.array(image_file))
print(np.array(Image.open('./img/eren_rgb_bin.png')))
```

```python
from IPython.display import display
from PIL import Image
import numpy as np

image_file = Image.open("./img/eren_rgb.png") # open colour image
image_file = image_file.convert('1', dither=Image.NONE) # convert image to black and white
image_file.save('./img/eren_rgb_bin1.png')
display(image_file)
print(np.array(image_file))
print(np.array(Image.open('./img/eren_rgb_bin1.png')))
```
